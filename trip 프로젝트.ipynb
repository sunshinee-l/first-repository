```python
from datetime import datetime

data['pickup'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['dropoff'] = pd.to_datetime(data['tpep_dropoff_datetime'])
data['diff'] = (data['dropoff'] - data['pickup']).dt.total_seconds()/60
data['diff'] = data['diff'].round(2)
data['diff']
```

```python
data['trip_hour'] = data['diff']
data
```

```python
sns.scatterplot(x= data['fare_amount'] , y=data['trip_distance'])
```

```python
sns.scatterplot(x= data['fare_amount'] , y=data['trip_hour'])
```
