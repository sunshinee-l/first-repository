```python
import random
class Character:
    def __init__(self,name):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack #공격력
        self.defense = defense #방어력
        
    def is_alive(self):
        return self.hp > 0
    
    def take_damage(self, damage):
        if damage > self.defense:
            self.hp -= (damage - self.defense)
        else:
            print()
            
    def attack_target(self,target):
            damage = random.randint(1, self.attack)
            target.take_damage(damage)
            
def battle(player, monster):
    print(f"\n{monster.name}과의 전투를 시작합니다.")
    while player.is_alive() and monster.is_alive():
        damage = random.randint(1, player.attack)
        print(f"{player.name}이 {monster.name}에게 {damage} 만큼 공격했다...!")
        monster.take_damage(damage)
        print(f"{monster.name}의 체력: {monster.hp}")
        if not monster.is_alive():
            break
            
        damage = random.randint(1, monster.attack)
        print(f"{monster.name}이 {player.name}에게 {damage} 만큼 공격했다...!")
        player.take_damage(damage)
        print(f"{player.name}의 체력: {player.hp}")

    if  player.is_alive():
        player.gain_experience(monster.level * 20)
        player.level_up()
        print("전투 승리!")
    else:
        print("전투 패배...")
            
class Player(Character):
    def __init__(self,name):
        self.name = name
        self.level = 1
        self.hp = 100
        self.attack = 25 #공격력
        self.defense = 5 #방어력
        self.experience = 0
        
    def gain_experience(self,gain):
        self.experience += gain
        
    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
        else:
            pass
            
class Monster(Character):     
    def __init__(self,name,level):
        self.name = name
        self.level = level
        self.hp = random.randint(10, 30) * self.level
        self.attack = random.randint(5, 20) * self.level
        self.defense = random.randint(1,5) * self.level
        

def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    player_name = input("이름을 입력해주세요: ")   # 이름 입력
    player = Player(player_name)
    
    for name, level in monster_dict.items():
        monster = Monster(name, level) 
        battle(player, monster)

        if not player.is_alive():
            print("게임 오버")
            break 

main()
```
