```python
import random
class Account:
    account_count = 0
    def __init__(self):
        self.bank_name = "SC은행"
        owner = input("예금주를 입력해주세요:")
        self.owner = owner
        
        self.account_number = (str(random.randint(100, 999))+'-'+str(random.randint(10, 99))+'-'+str(random.randint(100000, 999999)))
        
        balance = input("초기잔액을 입력해주세요:")
        self.balance = int(balance)
        
        self.deposit_log = []
        self.withdraw_log = []
        
        print(f"은행이름: {self.bank_name}, 계좌번호: {self.account_number}")
        Account.account_count += 1  # 계좌 개수 증가
    
    @classmethod
    def get_account_num(cls):  # ← 여기서 cls는 class Account를 가리켜!
        return cls.account_count
    
    def deposit(self):
        self.deposit_count = 0
        self.total_deposit = 0
        amount = int(input("입금하려는 금액을 적어주세요: "))
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_log.append(f"{amount}원 입금됨")
      
            self.total_deposit += amount
            print(f"입금금액은 {amount}입니다. 총 현재 금액은 {self.balance}")
        else:
            print("금액을 다시 적어주세요.")
        #추가 입금
        while amount != 0:
            print()
            amount = int(input("입금하려는 금액을 적어주세요: "))
            if amount <= 0:
                print("입금을 종료합니다.")
                break
            
            self.balance += amount
            self.deposit_count += 1
            self.deposit_log.append(f"{amount}원 입금됨")
            self.total_deposit += amount
            
            print(f"입금금액은 {amount}입니다. 총 현재 금액은 {self.balance}")
            if self.deposit_count >= 5:
                self.balance *= 1.01 
                print(f"이자가 추가되었습니다. 현재 잔액: {self.balance}")
            
    def withdraw(self):
        self.total_withdraw = 0
        amount = int(input("출금하시려는 금액을 입력해주세요. : "))
        if self.balance >= amount:
            self.balance -= amount
            
            self.withdraw_log.append(f"{amount}원 입금됨")
            self.total_withdraw += amount
            
            print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.balance}원")
        else:
            print("출금 실패: 잔액이 부족합니다.")
            
    def display_info(self):
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance}")
    
    def deposit_history(self):
        if not self.deposit_log:
            print("입금 내역이 없습니다.")
        else:
            print(f"입금 내역:{self.total_deposit}")
    
    def withdraw_history(self):
        if not self.withdraw_log:
            print("입금 내역이 없습니다.")
        else:
            print(f"출금 내역:{self.total_withdraw}")
        
    
a = Account()
a.deposit() 
a.withdraw()
a.display_info()
a.deposit_history()
a.withdraw_history()

accounts = []  # Q8: 여러 계좌 저장할 리스트
# Q8: 계좌 여러 개 생성해서 리스트에 저장
for i in range(3):
    acc = Account()
    accounts.append(acc)
    
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()  
```
